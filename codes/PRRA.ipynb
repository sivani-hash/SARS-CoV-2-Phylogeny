```python
from Bio import AlignIO

alignment = AlignIO.read("data/clustalw.aln", "clustal")

print("--- Furin Cleavage Site Alignment --- \n")

mutation_site = 0
for record in alignment:
    if "YP_009724390.1" in record.id: 
        mutation_site = record.seq.find("PRRA")

for record in alignment:
    protein_window = record.seq[mutation_site - 5 : mutation_site + 10]
    
    print(f"{record.id:15} | {protein_window}")
```
