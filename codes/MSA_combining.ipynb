```python
from Bio import SeqIO

# Import fasta seq from data folder
input_files = [
    "data/sars_cov_2.fasta", 
    "data/sars_cov_1.fasta", 
    "data/ratg13.fasta", 
    "data/mers.fasta", 
    "data/banal_236.fasta"
]
# Combine all seq and save to data folder
output_file = "data/combined.fasta"

# 2. Parse and combine records
combined_records = []
for file in input_files:
    # Read each file and append sequence records to our list
    for record in SeqIO.parse(file, "fasta"):
        combined_records.append(record)

# 3. Write all sequences into the single output file
with open(output_file, "w") as out_handle:
    SeqIO.write(combined_records, out_handle, "fasta")

# 4. Fixed the missing closing parenthesis here!
print(f"Successfully combined {len(input_files)} files into {output_file}!")
```
